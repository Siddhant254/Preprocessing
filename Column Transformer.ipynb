## COLUMN TRANSFORM

- Helps to transform a column without splitting the original data .

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Users\\siddh\\OneDrive\\Desktop\\Data science data\\Preprocessing\\covid_toy.csv")
df.head()
```

```python
# NULL VALUES
df.isnull().sum()
```

```python
# GENDER - NOMINAL CATEGORICAL DATA --> ONE HOT ENCODING
# FEVER - NUMERICAL DATA WITH MISSING VALUES --> SIMPLE IMPUTER
# COUGH - ORDINAL CATEGORICAL DATA --> ORDINAL ENCODER
# CITY - NOMINAL CATEGORICAL DATA --> ONE HOT ENCODING
# HAS_COVID - TARGET VARIABLE --> LABEL ENCODING
```

```python
# TRAIN-TEST SPLIT
from sklearn.model_selection import train_test_split
X = df.drop(columns=['has_covid'])
y = df['has_covid']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
X_train.head()
```

```python
# IMPORTING COLUMN TRANSFORMER
from sklearn.compose import ColumnTransformer
```

```python
# IMPORTING REQUIRED CLASSES FOR TRANSFORMATION
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder
from sklearn.preprocessing import OneHotEncoder
```

```python
# CREATING TRANSFORMER OBJECT
transformer = ColumnTransformer(transformers=[
    ('trf1',SimpleImputer(),['fever']),
    ('trf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('trf3',OneHotEncoder(sparse=False,drop = 'first'),['gender','city'])
],remainder='passthrough')
```

```python
# FITTING TRAINING DATA TO TRANSFORMER
transformer.fit(X_train)
```

```python
# TRANSFORMING TRAINING AND TEST DATA
X_train_transformed = transformer.transform(X_train)
X_test_transformed = transformer.transform(X_test)
```

```python
# CONVERTING TO DATAFRAME
X_train_transformed = pd.DataFrame(X_train_transformed , columns = ['fever','cough','gender',
                                                                   'Delhi','Kolkata','Mumbai','age'])
X_test_transformed = pd.DataFrame(X_test_transformed,columns=['fever','cough','gender',
                                                     'Delhi','Kolkata','Mumbai','age'])
```
