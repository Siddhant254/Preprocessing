## Working with dates

```python
orders.info()
```

```python
# CONVERTING TO DATETIME FORMAT
orders['date'] = pd.to_datetime(orders['date'])
```

```python
orders.info()
```

### 1. Extract year from date

```python
orders['year'] = orders['date'].dt.year
orders.head()
```

### Extract Month_no. from date

```python
orders['month_no'] = orders['date'].dt.month
orders.head()
```

### Extract month name from date

```python
orders['month_name'] = orders['date'].dt.month_name()
orders.head()
```

### Extract day no. of month from date

```python
orders['day_no_month'] = orders['date'].dt.day
orders.head()
```

### Extract day no.  of the week from date

```python
orders['day_no_week'] = orders['date'].dt.dayofweek
orders.head()
```

### Extract day of the week name

```python
orders['day_name_week'] = orders['date'].dt.day_name()
orders.head()
```

### Checking for the weekend

```python
orders['is_weekend'] = np.where(orders['day_name_week'].isin(['Saturday','Sunday']),1,0)
orders.iloc[:,4:].head()
```

### Extract week of the year

```python
orders['week_of_year'] = orders['date'].dt.week
orders.iloc[:,4:].head()
```

### Extract quarter

```python
orders['quarter_of_year'] = orders['date'].dt.quarter
orders.iloc[:,-2:].head()
```

#### Extract time elapsed between dates

```python
import datetime
today = datetime.datetime.today()

(today - orders['date']).dt.days
```
